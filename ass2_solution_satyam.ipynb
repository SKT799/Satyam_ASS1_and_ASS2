```python
import pandas as pd
```

```python
# Load the Data into Google Sheets
df = pd.DataFrame(dataset['train'])
df.head()
```

```python
df
```

```python
# Hindi and English Dataset spliting:
midpoint =len(df)//2
# midpoint

english_data = df['text'].iloc[:midpoint].reset_index(drop=True)
# english_data

hindi_data = df['text'].iloc[midpoint:].reset_index(drop=True)
# hindi_data

processed_df = pd.DataFrame({
    #coloum1
    'English Sentences':english_data,

    #coloum2
    'Hindi Sentences':hindi_data
}
)

processed_df
```

```python
# Keep only those words count whose sentences fall within the range of 5 to 50 in both languages.

#count english:
processed_df['Word Count(English)'] = processed_df['English Sentences'].astype(str).str.split().str.len()



#count hindi:
processed_df['Word Count(Hindi)'] = processed_df['Hindi Sentences'].astype(str).str.split().str.len()


# processed_df

#Filter English:
processed_df = processed_df[(processed_df['Word Count(English)']>=5) & (processed_df['Word Count(English)'] <=50)]

#Filter Hindi:
processed_df = processed_df[(processed_df['Word Count(Hindi)']>=5) & (processed_df['Word Count(Hindi)'] <=50)]

processed_df
```

```python
# Retain only those word pairs where the word count difference lies within the range of -10 to +10.

processed_df['Difference between Word Count(English) and Word Count(Hindi)'] = processed_df['Word Count(English)'] - processed_df['Word Count(Hindi)']


processed_df = processed_df[(processed_df['Difference between Word Count(English) and Word Count(Hindi)']>=-10) & (processed_df['Difference between Word Count(English) and Word Count(Hindi)']<=10)]

processed_df
```

```python
# Final Output:
'''
English Sentences
Hindi Sentences
Word Count (English)
Word Count (Hindi)
Difference between Word Count (English) and Word Count (Hindi)

'''

final_coloums =[
    'English Sentences',
    'Hindi Sentences',
    'Word Count(English)',
    'Word Count(Hindi)',
    'Difference between Word Count(English) and Word Count(Hindi)'
]

final_df = processed_df[final_coloums]
# final_df

final_df.to_excel("iitp_nlp_satyam.xlsx")
```
